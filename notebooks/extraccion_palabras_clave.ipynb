```python
# Carga de archivos para la limpieza y preprocesamiento
from google.colab import files
import os

print("Por favor, sube tu archivo txt o docx")
texto_subido = files.upload()

contenido = ""

if texto_subido: # Check if any file was uploaded
    # Obtención del nombre del archivo del diccionario de files.upload()

    nombre_archivo = list(texto_subido.keys())[0]
    file_extension = os.path.splitext(nombre_archivo)[1].lower()

    if file_extension == ".docx":
        !pip install docx2txt # Ensure docx2txt is installed
        import docx2txt

        docx_file = nombre_archivo
        txt_file = nombre_archivo.replace('.docx', '.txt')

        # Convertir docx a texto
        contenido = docx2txt.process(docx_file)

        # Guardar el texto en un archivo .txt
        with open(txt_file, "w", encoding="utf-8") as f:
            f.write(contenido)
        print(f"Archivo '{nombre_archivo}' (docx) convertido y guardado como '{txt_file}'.")
    elif file_extension == ".txt":
        with open(nombre_archivo, "r", encoding="utf-8") as archivo:
            contenido = archivo.read()
        print(f"Archivo '{nombre_archivo}' (txt) cargado.")
    else:
        print(f"Tipo de archivo '{file_extension}' no soportado. Por favor, sube un archivo .txt o .docx.")
else:
    print("Ningún archivo fue subido.")

```

```python
#Uso STOP_WORDS de Spacy en español
from spacy.lang.es import STOP_WORDS
stop_words = STOP_WORDS

# Establecer STOP WORDS
spanish_stopwords = list(STOP_WORDS)

#extracción de palabras clave por sklearn y tfidftransformer
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer



def palabras_clave_TFIDF (texto):
  #inicializar variable df
  df = texto
  #hacer el recuento vectorial de las palabras
  cv = CountVectorizer(stop_words=spanish_stopwords, max_df=1.0, min_df=1)
  word_counts = cv.fit_transform([df])
  #obtener lista de palabras en los documentos
  features = cv.get_feature_names_out()
  #imprime las palabras de los documentos
  print (features)
  transformer = TfidfTransformer ()
  transformer.fit(word_counts)
  count_vector = cv.transform ([contenido])

  #vectores de las palabras
  tfidf_vector = transformer.transform(count_vector).tocoo()
  tuples = zip (tfidf_vector.row, tfidf_vector.col, tfidf_vector.data)
  tfidf_vector = sorted (tuples, key=lambda x: x [2], reverse=True)

  #obtiene las palabras claves
  top_keywords = tfidf_vector [:5]
  for tup in top_keywords:
    print (features [tup [1]], [tup [2]],)

palabras_clave_TFIDF (contenido)
```
